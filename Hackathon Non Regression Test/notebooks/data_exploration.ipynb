# Dataset Exploration

Notebook esplorativo per comprendere la struttura del dataset e le sue caratteristiche principali.

```python
import os
import json
import boto3
from dotenv import load_dotenv
load_dotenv()

S3_BUCKET = os.environ["S3_BUCKET"]
AWS_REGION = os.environ.get("AWS_REGION", "eu-west-1")
AWS_ACCOUNT_ID = os.environ.get("AWS_ACCOUNT_ID", "")
DATASET_KEY = "dataset/dataset.jsonl"

s3 = boto3.client("s3", region_name=AWS_REGION)
print(f"Bucket: {S3_BUCKET} | Region: {AWS_REGION}")
```

```python
# Caricamento dataset da S3
_owner_kw = {"ExpectedBucketOwner": AWS_ACCOUNT_ID} if AWS_ACCOUNT_ID else {}
response = s3.get_object(Bucket=S3_BUCKET, Key=DATASET_KEY, **_owner_kw)
lines = response["Body"].read().decode("utf-8").splitlines()
records = [json.loads(l) for l in lines if l.strip()]

print(f"Record totali: {len(records)}")
```
